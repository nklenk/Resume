# Visualizing Geographic Data

### Below I use the Basemap library to create some geographic plots of information reguaring airports, airlines, and some of the most popular flightpaths



```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

airlines = pd.read_csv('//Users//neilklenk//Documents//Datasets//Airplanes//airlines.csv')
airports = pd.read_csv('//Users//neilklenk//Documents//Datasets//Airplanes//airports.csv')
routes  = pd.read_csv('//Users//neilklenk//Documents//Datasets//Airplanes//routes.csv')
```

Since two dimensional maps are easier to understand and less computationally intensive, the latitude and
Longitude values need to be changed to the cartesian coordinate system using a Mercator projection
Latitude and longitude values are also used to specify which portion of the map that we want to use

```python
fig, ax = plt.subplots(figsize=(7,7))
plt.title("Airport Locations Around the World")
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180)
longitudes = airports["longitude"].tolist()
latitudes = airports["latitude"].tolist()
# mapping the longitude and latitude values onto the cortesian coordinate system via a Mercator projection
x, y = m(longitudes, latitudes)
# plot the new coortesian coordinate values
m.scatter(x, y, s=1)
# Add the coastline to give the viewer a reference
m.drawcoastlines()
plt.show()
```

Great circles will be used to better understand the flight paths that are taken by the planes.
A great circle is demonstrated as a shortest circle connecting two points, and this visualizes the flight paths

```python
geo_routes = pd.read_csv("//Users//neilklenk//Documents//Datasets//Airplanes//geo_routes.csv")

fig, ax = plt.subplots(figsize=(7,7))
plt.title("Flight Paths Leaving DFW")
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180)
m.drawcoastlines()
```

A function was written to handle the creation of the great circles. It limits the values to those within our map projection

```python
def create_great_circles(df):
    for index, row in df.iterrows():
        end_lat, start_lat = row['end_lat'], row['start_lat']
        end_lon, start_lon = row['end_lon'], row['start_lon']
        
        if abs(end_lat - start_lat) < 180:
            if abs(end_lon - start_lon) < 180:
                m.drawgreatcircle(start_lon, start_lat, end_lon, end_lat)
```

Visualizing all of the flightpaths that originate from the DFW airport

```python
dfw = geo_routes[geo_routes['source'] == "DFW"]
create_great_circles(dfw)
plt.show()
```
